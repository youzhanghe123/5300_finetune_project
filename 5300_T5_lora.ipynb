# Create dataset

```python
# Load the Wikipedia Summary dataset (e.g., English subset for text summarization)
# Load the Wikipedia Summary dataset (e.g., English subset for text summarization)
from datasets import load_dataset

# Load the CNN/DailyMail dataset for summarization
dataset = load_dataset("billsum", split="ca_test")
```

```python
dataset = dataset.train_test_split(test_size=0.2)
```

```python
# Split into train and test datasets
train_dataset = dataset["train"].select(range(900))  # Subset for demonstration (1000 samples)
test_dataset = dataset["test"].select(range(200))    # Subset for demonstration (200 samples)
```

# Load original model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
from transformers import Trainer, TrainingArguments

from datasets import load_dataset
import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python

tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
model
```

```python
import evaluate
import numpy as np

# Load ROUGE metric
rouge = evaluate.load("rouge")
```

```python
# Updated tokenize function
def tokenize_function(examples):
    inputs = [doc for doc in examples["text"]]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer([str(summary) for summary in examples["summary"]],
                         max_length=128,
                         truncation=True,
                         padding="max_length")

    model_inputs["labels"] = labels["input_ids"]


    return model_inputs

```

```python
# Apply tokenization to the dataset
tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)
```

```python
import torch

# Clear CUDA memory cache
torch.cuda.empty_cache()
```

# Supervised fine tune

```python
# Define the compute_metrics function
def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    # Decode the predictions and labels (token IDs) into text
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Rouge expects a list of predictions and references
    rouge_results = rouge.compute(predictions=decoded_preds, references=decoded_labels)

    # Extract relevant scores
    return {
        "rouge1": rouge_results["rouge1"],
        "rouge2": rouge_results["rouge2"],
        "rougeL": rouge_results["rougeL"],
    }
```

```python
# Assuming `model` is your fine-tuned model (e.g., BartForConditionalGeneration)

# Calculate total trainable parameters
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Total trainable parameters: {trainable_params}")
```

```python
# 1. Load the pre-trained model and tokenizer directly
from peft import get_peft_model, LoraConfig, TaskType, prepare_model_for_kbit_training
import torch
from datasets import Dataset

# 2. Configure LoRA
# Define LoRA configuration for DistilBART (Sequence-to-Sequence)
lora_config = LoraConfig(
    task_type=TaskType.SEQ_2_SEQ_LM,  # Sequence-to-sequence task
    r=8,  # Rank for low-rank approximation
    lora_alpha=32,  # Scaling factor
    lora_dropout=0.1,  # Dropout rate
    bias="none",
    # For DistilBART, target specific attention and feed-forward layers
    target_modules=[
        "SelfAttention.q",  # Query projection in self-attention
        "SelfAttention.v",  # Value projection in self-attention
        "SelfAttention.k",  # Output projection in self-attention
        "EncDecAttention.q",  # First linear layer in feed-forward network
        "EncDecAttention.k",
        "EncDecAttention.v"# Second linear layer in feed-forward network
    ],
    modules_to_save=["lm_head"],  # Save the language modeling head
)


# Prepare the model for low-rank adaptation with LoRA
model = prepare_model_for_kbit_training(model)
peft_model = get_peft_model(model, lora_config).to(device)

# Print trainable parameters
print("Trainable parameters:")
peft_model.print_trainable_parameters()
```

```python
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments, DataCollatorForSeq2Seq
from datasets import load_dataset
```

```python
# Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=5,
    predict_with_generate=True,  # Enables generation for ROUGE scoring
    logging_dir="./logs",
    save_strategy="epoch",
    load_best_model_at_end=True,  # Loads the best model for evaluation
    metric_for_best_model="rouge2",  # Track ROUGE-2 for the best model

    logging_strategy="steps",  # Log by steps for continuous logs
    logging_steps=10,  # Set desired logging frequency
)
```

```python
trainer = Seq2SeqTrainer(
    model=peft_model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```
